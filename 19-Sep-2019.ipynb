```python
import pandas as pd
```

```python
census_df=pd.read_csv('census.csv')
census_df.head()
```

```python
import numpy as np
def min_max(row):
    data=row[["POPESTIMATE2010",
             "POPESTIMATE2011",
             "POPESTIMATE2012",
             "POPESTIMATE2013",
             "POPESTIMATE2014",
             "POPESTIMATE2015"]]
    return pd.Series({'min':np.min(data),'max':np.max(data)})
```

```python
census_df[['min','max']]=census_df.apply(min_max,axis=1)
```
